### Geração do PDF de Qr-Code

```python
import requests
from fpdf import FPDF
import os
# save FPDF() class into a
# variable pdf
pdf = FPDF()

# set style and size of font
	# that you want in the pdf
pdf.set_font("Arial", size = 15)



for index,i in enumerate(dfGeral.values.tolist()):
    sala = i[0] #nome da sala/pavimento

    pdf.add_page() # add a page
    
	# create a cell
    pdf.cell(w = 0, h = 100, txt = sala,
			ln = 1, align = 'C')

    

    
    response = requests.get(f"https://us-central1-eficiencia-energetica---trt.cloudfunctions.net/api/salasQrCode?sala={sala}").json()
    qrCodeSala = response["qrCodeSala"]
    response2 = requests.get(qrCodeSala)

    imgDir = f"C:\\Users\\joaom\\OneDrive\\Área de Trabalho\\TRT - PROJETO\\{index}.png"

    file = open(imgDir, "wb")
    file.write(response2.content)
    file.close()

    #BytesIO(response2.content)
    pdf.image(imgDir, x = pdf.w/2-50/2, y = pdf.h/2-50/2, w = 50, h = 50)

    os.remove(imgDir)#remove a imagem da pasta

    break

pdf.output(r"C:\Users\joaom\OneDrive\Área de Trabalho\TRT - PROJETO\QrCode-Salas.pdf")
```

### Carga no Banco de Dados

```python
import requests
for i in dfGeral.values.tolist():
    json = {"sala":i[0],"comprimento":i[1],"largura":i[2],"distanciaPlanoDeTrabalhoTeto":i[3],"potenciaLampada":i[4],"qntdLampadas":i[5]}
    response = requests.post(f"https://us-central1-eficiencia-energetica---trt.cloudfunctions.net/api/salas",json=json)
    print(response.status_code)
```
